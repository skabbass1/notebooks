```python
import numpy as np
import pandas as pd 
```

```python
dates = pd.date_range("20100101", "20190901")
asset_prices = {
    "AAPL": np.random.uniform(low=100, high=120, size=len(dates)),
    "GOOG": np.random.uniform(low=80, high=100, size=len(dates)),
    "MSFT": np.random.uniform(low=200, high=210, size=len(dates)),
    "PYPL": np.random.uniform(low=250, high=300, size=len(dates))
}
df = pd.DataFrame(asset_prices, index=dates)
```

# Multi-line charts 

```python
# Additional examples (including handling large data) at https://plot.ly/python/line-and-scatter/
import plotly.graph_objects as go

fig = go.Figure(data=go.Scatter(x=df.index, y=df.AAPL, mode='lines', name="AAPL"))
fig.add_trace(go.Scatter(x=df.index, y=df.GOOG, mode='lines', name="GOOG"))
fig.add_trace(go.Scatter(x=df.index, y=df.GOOG, mode='markers', name="MSFT"))
```

# Subplots

```python
# more examples here:https://plot.ly/python/subplots/

from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=2, cols=2, subplot_titles=("GOOG", "MSFT", "AAPL", "PYPL "))


fig.add_trace(go.Scatter(x=df.index, y=df.GOOG, mode='lines', name="GOOG"), row=1, col=1)
fig.add_trace(go.Scatter(x=df.index, y=df.MSFT, mode='lines', name="MSFT"), row=1, col=2)
fig.add_trace(go.Scatter(x=df.index, y=df.AAPL, mode='lines', name="AAPL"), row=2, col=1)
fig.add_trace(go.Scatter(x=df.index, y=df.PYPL, mode='lines', name="PYPL"), row=2, col=2)

fig.update_layout(
    title=go.layout.Title(
        text="Title",
        xanchor="left",
        yanchor="auto",
    ))
fig.show()
```

# Mark Regions on Chart

```python
# more info here: https://plot.ly/python/shapes/

import plotly.graph_objects as go
import numpy as np
from plotly.subplots import make_subplots


x = np.arange(10)

fig = make_subplots(2,1)
fig.add_trace(go.Scatter(x=x, y=x**2), row=1, col=1)
fig.add_trace(go.Scatter(x=x, y=x**3), row=2, col=1)
fig.update_layout(
    shapes=[
        # Line reference to the axes
        go.layout.Shape(
            type="rect",
            xref="x2",
            yref="y2",
            x0=4,
            y0=0,
            x1=6,
            y1=800,
           
             fillcolor="LightSalmon",
            opacity=0.5,
            layer="below",
            line_width=0,
          
        ),
    ]
)


fig.show()
```
